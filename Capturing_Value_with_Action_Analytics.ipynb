```python
from googleapiclient.discovery import build

api_key = 'AIzaSyB4wmL22JgfbEnp24whx9QhILcRtPuGcvk'

# Create a YouTube Data API client
youtube = build('youtube', 'v3', developerKey=api_key)
```

```python
video_id = 'EQoJpMjMiG8'
video_data = youtube.videos().list(part='snippet,statistics', id=video_id).execute()

# Extract relevant information
video_info = video_data['items'][0]['snippet']
statistics = video_data['items'][0]['statistics']

# Display video title, view count, and likes
print('Video Title:', video_info['title'])
print('View Count:', statistics['viewCount'])
print('Likes:', statistics['likeCount'])
```

```python
comments = youtube.commentThreads().list(part='snippet', videoId=video_id, textFormat='plainText').execute()

# Display comments
for comment in comments['items']:
    text = comment['snippet']['topLevelComment']['snippet']['textDisplay']
    print('Comment:', text)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with video information
data = {
    'Video Title': [video_info['title']],
    'View Count': [int(statistics['viewCount'])],
    'Likes': [int(statistics['likeCount'])]
}
df = pd.DataFrame(data)

# Visualize data
df.plot(kind='bar', x='Video Title', y=['View Count', 'Likes'])
plt.title('Video Analytics')
plt.ylabel('Count')
plt.show()
```
